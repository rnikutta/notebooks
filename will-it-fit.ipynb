##### Imports

```python
import willitfit as W
%matplotlib inline
```

##### Load model fluxes (N-dim hypercube) and list of parameter vectors

```python
fluxes, theta = W.grappa_get_data_theta(npzfile='/home/robert/science/data/grappa/models/newLibNSFgals/grasil_test_003.npz')
```

##### theta vectors (unique sampled parameter values)

```python
theta
```

##### Shape of fluxes hypercube

```python
fluxes.shape
```

##### Instatiate a database object which provides methods to interpolate model fluxes

```python
db = W.DB(fluxes,theta)
```

##### Load observed data of one source

```python
source = W.Source('/home/robert/science/grappa/data/smallPEP/rest_obj_15271_z0.955.dat')
```

##### Finally, instatiate a model class. This provides methods for MCMC sampling etc.

```python
model = W.Model(db,source,samplesize=1e5,burn=2e4)
```

##### Run the sampler, compute statistics

```python
model.analyze()
```

##### Make an overview plot

```python
pl = W.plotall(model)
```

##### Write results (best-fit SEDs, stats, tabulations of posteriors, etc.) to ASCII files

```python
W.output(model)
```
